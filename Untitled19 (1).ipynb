```python
from __future__ import print_function
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD
from keras.utils import np_utils
import pandas as pd
from keras import optimizers
```

```python
# read exam.csv into x features, y for classifications
# Code to read data into x and y goes here
data = pd.read_csv("exam.csv", header=None)
data.insert(0,'bias',1) #intercept
cols = data.shape[1] #number of columns
x = data.iloc[1:,0:cols-1] #features
y = data.iloc[1:,cols-1:cols] #target (result)
x = np.array(x.values, dtype=float)
y = np.array(y.values, dtype=float)

#specify number of iterations in NB_EPOCH
NB_EPOCH = 400
 

# specify number of hidden layers
N_HIDDEN = 30

 
BATCH_SIZE = 2

 

VERBOSE = 1

 

# specify number of classifications

NB_CLASSES = 2

 

VALIDATION_SPLIT=0.1 # how much TRAIN is reserved for VALIDATION

#

examXTrain = x[0:80,0:2]
examYTrain = y[0:80]
examXTest  = x[0:80:,0:2]
examYTest  = y[0:80]

#

X_train = examXTrain.reshape(80, 2)
X_test = examXTest.reshape(80, 2)

#

X_train = examXTrain
X_test  = examXTest
y_train = examYTrain
y_test  = examYTest
Y_train = np_utils.to_categorical(y_train, NB_CLASSES)
Y_test  = np_utils.to_categorical(y_test, NB_CLASSES)

#

model = Sequential()
model.add(Dense(N_HIDDEN, kernel_initializer='uniform', input_shape=(2,)))
model.add(Activation('relu'))
 
#

model.add(Dense(N_HIDDEN))
model.add(Activation('relu'))
model.add(Dense(NB_CLASSES))

#

model.add(Activation('softmax'))
model.summary()

#change the learning rate

sgd = optimizers.SGD(lr=0.001, decay=1e-6, momentum=0.0, nesterov=True)

#'mean_squared_error'categorical_crossentropy

model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

#

history = model.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=NB_EPOCH, verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
score = model.evaluate(X_test, Y_test, verbose=VERBOSE)

 

print("Test score:", score[0])

print('Test accuracy:', score[1])

 

#Predict

#to predict create a 2-dim array with grades and invoke
grade = np.array([[87.5,90],[27.5,29.0],[7.5,19],[67.5, 89.0]])
model.predict_classes(grade)
```
